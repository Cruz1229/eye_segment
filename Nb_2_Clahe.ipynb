```python
from dr_filters.clahe  import clahe
import cv2
import os
```

```python
import json

with open("config.json") as json_config_file:
    config_file = json.load(json_config_file)
```

```python
img_clahe = clahe.equalize(image_file='./output/seqnet/impaint/40_left.png', gray_scale=False, with_plot=True)


output_path = config_file["clahe"]["output"]
os.makedirs(f"{output_path}", exist_ok=True)
cv2.imwrite(f"{output_path}/40_left.png", cv2.cvtColor(img_clahe, cv2.COLOR_RGB2BGR))
```
